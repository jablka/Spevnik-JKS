```python
import pandas as pd
from bs4 import BeautifulSoup
import copy
```

```python
s = '''
<!DOCTYPE html>
<html lang="sk">
<head>
    <meta charset="UTF-8">
    <meta content="width=device-width, initial-scale=1.0, user-scalable=yes" name="viewport"/>    
    <link href="../moj.css" rel="stylesheet">
    <script src="../moj.js"></script>
</head>
<body>
    <script>init();</script>
<div id="pies_cont"></div>
</body>
</html>
'''

soup = BeautifulSoup(s, 'lxml')
```

```python
df = pd.read_csv('texty_s_nbsp.csv', index_col=0)
df
```

```python
# pozor, cislo je string, sort musí byť False!
df_grouped = df.groupby(['cislo'], sort=False)[['strofa','text']].apply(lambda x:x)
df_grouped
```

```python
# pozor, cislo je string, sort musí byť False!
gg = df.groupby(['cislo'], sort=False)[['strofa','text']]
gg
```

```python
for number, song in gg:
    with open(f'C:\\piesne202\\res\\{number[0]}.html', mode='w', encoding='utf-8') as f_out:   

        soup_file = copy.copy(soup)

        my_string = f'<div id="cislo">» {str(number[0])} «</div>' # »«
        
        my_string += '<div>'
        for row in song.itertuples():
            # rr = row[2].replace(' - ', '\n') 
            rr=row[2]
            strofa = f"<span class='strofa'>{str(row[1])}. </span>"
            # line = str(row[1])+'. '+rr+'<br>'
            line = '<p>'+strofa+rr+'</p>'
            my_string += line
            # my_string += '<br>'
        my_string += '</div>'
        
        my_string += "<div class='ending'>☙</div>"
        content = BeautifulSoup(my_string,'html.parser')
        soup_file.div.append(content)

        f_out.write(str(soup_file))
```
