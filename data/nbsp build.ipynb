```python
import re
import regex # third party package
```

```python
with open('texty.csv', encoding='utf-8') as f:
    s = f.read()
```

checking and replacing, first round:

```python
pat = r'\b(\w{1})\b \b(\w{1,})\b'
vysl = re.findall(pat,s)
print(len(vysl))
print(vysl)
```

```python
# checking with third party package:
vysl = regex.findall(pat,s,overlapped=True) # overlapped=True
len(vysl)
```

```python
# replacing
repl = f'\\1\u00a0\\2' # \u00a0 non-breaking space
s,n = re.subn(pat, repl, s)
# print(s)
print(n)
```

checking and replacing, second time:

```python
pat = r'\b(\w{1})\b \b(\w{1,})\b'
vysl = re.findall(pat,s)
print(len(vysl))
print(vysl)
```

```python
# checking with third party package:
vysl = regex.findall(pat,s,overlapped=True)
len(vysl)
```

```python
# replacing
repl = f'\\1\u00a0\\2' # \u00a0 non-breaking space
s,n = re.subn(pat, repl, s)
# print(s)
print(n)
```

final checking:

```python
pat = r'\b(\w{1})\b \b(\w{1,})\b'
vysl = re.findall(pat,s)
len(vysl)
```

```python
vysl = regex.findall(pat,s,overlapped=True)
len(vysl)
```

saving to file:

```python
with open('texty_s_nbsp.csv', 'w', encoding='utf-8') as f:
    f.write(s)
```
